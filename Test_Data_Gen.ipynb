## Food

```python
foods = ['Chole Bhature', 'Cholar Dal', 'Dosa', 'Dhokla', 'Hummus', 'Sausage']

for food in foods:
    for word in ambiguos_words:
        test_sentences.append(f"My {word} is cooking {food} now.")
        test_sentences.append(f"My {word} is eating {food}  for lunch.")
        test_sentences.append(f"My {word} served {food} at the family dinner.")
        test_sentences.append(f"Their {word} is cooking {food} now.")
        test_sentences.append(f"Their {word} is eating {food}  for lunch.")
        test_sentences.append(f"Their {word} served {food} at the family dinner.")

print(len(test_sentences))
```

## Location

```python
locations = ['village', 'city', 'town', 'country', 'continent']
loc_names = ['Kadua', 'Paradip', 'Bhubaneswar', 'Odisha', 'India', 'Ahmedabad', 'Gujarat', 'Africa', 'Europe', 'South America', 'Australia', 'Bengaluru', 'Kolkata']
for location in locations:
    for word in ambiguos_words:
        test_sentences.append(f"My {word} is in the {location}.")
        test_sentences.append(f"Their {word} is in the {location}.")
        test_sentences.append(f"My {word} is from a {location}.")
        test_sentences.append(f"Their {word} will go to a {location}.")

for loc_name in loc_names:
    for word in ambiguos_words:
        test_sentences.append(f"My {word} is in {loc_name}.")
        test_sentences.append(f"Their {word} is in {loc_name}.")
        test_sentences.append(f"My {word} is from {loc_name}.")
        test_sentences.append(f"Their {word} will go to {loc_name}.")

print(len(test_sentences))
```
