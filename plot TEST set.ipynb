```python
# make the y axis as the percentage instead of the count

def create_100_bars(data, title):
    df = pd.DataFrame(data).T.fillna(0)  # Transpose and replace NaN with zeros

    # Calculate percentage for each category within each language
    df_percentage = df.div(df.sum(axis=1), axis=0) * 100

    # Plotting
    ax = df_percentage.plot(kind='bar', stacked=True, color=[colours_map[cat] for cat in df_percentage.columns])

    # Adding value labels on bars
    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy() 
        if height <=34  and height >0:  # Avoid cluttering with zero-height patches (no data)
            ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')
 # Adding total items at the top of each bar
    totals = df.sum(axis=1)
    for i, total in enumerate(totals):
        ax.text(i, 102, f'{int(total)}', ha='center', va='bottom', fontsize=9)
 # add gap at the top of the y-axis
    plt.ylim(0, 100 * 1.1)

   # put legend outside of plot on the right side
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    # add custom y tick label  at height 102
    plt.text(-1, 102, "Total", ha='center', va='bottom', fontsize=9)

    
   
   

    # Setting labels and title
    plt.xlabel('Languages')
    plt.ylabel('Percentage')
    plt.title(title)
    # plt.legend(title='Categories')
    
    plt.tight_layout()
    
    # Save figure
    plt.savefig('results/new train set 100/gender_train_'+word+'.png')
    
    plt.show()

for word in df:
    create_100_bars(df[word], word)
```
