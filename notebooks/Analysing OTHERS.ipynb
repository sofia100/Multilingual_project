```python
import pandas as pd
import tqdm
import os
```

```python
dir = '../OTHERS'
# make a dataframe with columns - eng_text, tgt_lang, tgt_lang_text
df = pd.DataFrame(columns=['eng_text', 'tgt_lang', 'tgt_lang_text', 'eng_reln'])

for lang in languages:
    dir_path = f'{dir}/{lang}/'
    # read all csv files in the directory_path and make a dataframe 
    # with columns - hin_text, tgt_lang, tgt_lang_text
    for file in tqdm.tqdm(os.listdir(dir_path)):
        if file.endswith('.csv'):
            # read the csv file
            df1 = pd.read_csv(dir_path + file)
            # rename the column 'eng_Latn' to 'eng_text'
            df1.rename(columns = {'eng_Latn': 'eng_text'}, inplace=True)
            df1.rename(columns = {lang: 'tgt_lang_text'}, inplace=True) 
            # drop remarks if the column exists
            if 'remarks' in df1.columns:
                df1.drop(columns=['remarks'], inplace=True)
            # add the columns to the dataframe
            df1['tgt_lang'] = lang
            df1['eng_reln'] = file.split('_')[2]
            # append the dataframe to the main dataframe
            df = pd.concat([df, df1], ignore_index=True)
    
# save the dataframe to a csv file
df.to_csv('all_others.csv', index=False)

    
```
